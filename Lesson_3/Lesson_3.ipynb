#### Задание 1. В чем принципиальные отличия item-item рекомендаций от ALS?

item-item рекомендация - это детерминированный алгоритм, для каждого юзера находит ближайщих юзеров и на основе их скоров товаров дает скор для юзера, а ALS - это алгоритм матричнай факторизации, где раскладывается вся user-item матрица на матрицу латентных факторов юзеров и матрицу латентных факторов товаров и предсказание делается сразу для всех значений перемножением матриц.

#### Задание 2. Приведите 3 примера весов user-item матрицы для задачи рекоммендаций товаров

1) Количество покупок юзером данного товара  
2) Чистая прибыль с покупки юзером данного товара  
3) Метрика учитывающая несколько факторов (количестово покупок, комментарии, лайки и т.д.)  

#### Задание 3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т д)

1) Учитываются только взаимодействия юзера и товара, но не учитываются фичи юзеров или товаров .  
2) Никак не учитывается сезонность, праздники и т.д., т.е. алгоритм изменит скоры в матрице уже после того, как пользователи начнут взаимодействовать с товаром.  
3) Неприменим к юзерам или товарам у которых вообще нет взаимодействий.  
4) Результат работы алгоритма – это некоторые скоры, которые чем больше, тем лучше, а не шанс взаимодействия юзер/товар.
